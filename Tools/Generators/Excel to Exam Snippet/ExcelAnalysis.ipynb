# Processing into Question Data

```python
subjectName = input("Name of Subject: ")
paperName = input("Name of Paper: ")
```

```python
# Remove Escape Characters
def filtering(text):
    if len(text) <= 0: return text
    if text[-1] in ["\r","\n"]: return text[0:-1] 
    return text

index = {
    "subject": filtering(subjectName),
    "paper": filtering(paperName),
    "pdf":[],
    "questions": []
}

#Sample Question
sampleQuestion = {
    "subject": index["subject"],
    "paper": index["paper"],
    "q_id": "",
    "topics": [], #[]
    "images": [],
    "answerImages": [],
    "pdf": [],
    "anspdf": [],
    "question": "",
    "type": "open",
    "given": [""],
    "correct": [],
    "marks": [],
    "self_mark": True
  }
```

```python
import copy
def isNan(ans):
    return ans != ans

for i in range(df.shape[0]): #Number or rows
    #print("For Question ", i + 1)
    question = copy.deepcopy(sampleQuestion)
    question["q_id"] = str(i+1)
    row = df.iloc[i]
    
    question["question"] = row["Question"]
    for partans in range(maxAnsMark):
        ans = row[f"Answer {partans+1}"]
        if not isNan(ans): #Check for is not a number
            question["correct"].append(ans)
            question["marks"].append(1)
        else: break

    if not isNan(row["Type"]):
        question["topics"].append(row["Type"])
    if not isNan(row["Topic"]):
        question["topics"].append(row["Topic"])

    index["questions"].append(question)
    i+=1
```
